```python
import requests
```

```python
link='https://en.wikipedia.org/wiki/Languages_of_India'
```

```python
html=requests.get(link).text
```

```python
html
```

```python
bsobj=BeautifulSoup(html)
```

```python
bsobj.prettify()
```

```python
wiki_tbl=bsobj.find('table',class_='wikitable sortable')
```

```python
wiki_tbl
```

```python
tbl_links=wiki_tbl.find_all('a')
```

```python
tbl_links
```

```python
languages=[]
```

```python
for links in tbl_links:
  if links.get('title') is not None:

    print(links.get('title'))
```

```python
for links in tbl_links:
  if links.get('title') is not None:

     print(links.get('title').split(' ')[0])
     languages.append(links.get('title').split(' ')[0])
```

```python
languages
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(languages)
```

```python
df
```
